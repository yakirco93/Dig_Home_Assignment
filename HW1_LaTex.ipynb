## 1. Derivative of the Sigmoid Function

The sigmoid function is defined as:

$$
\sigma(x)=\frac{1}{1+e^{-x}}
$$

We can rewrite it as:

$$
\sigma(x)=(1+e^{-x})^{-1}
$$

Using the chain rule:

$$
\frac{d\sigma(x)}{dx}
=
-1(1+e^{-x})^{-2}\cdot(-e^{-x})
$$

Therefore:

$$
\frac{d\sigma(x)}{dx}
=
\frac{e^{-x}}{(1+e^{-x})^2}
$$

Now we show that this equals:

$$
\sigma(x)(1-\sigma(x))
$$

First:

$$
1-\sigma(x)
=
1-\frac{1}{1+e^{-x}}
$$

Using a common denominator:

$$
1-\sigma(x)
=
\frac{1+e^{-x}-1}{1+e^{-x}}
$$

So:

$$
1-\sigma(x)
=
\frac{e^{-x}}{1+e^{-x}}
$$

Now multiply:

$$
\sigma(x)(1-\sigma(x))
=
\frac{1}{1+e^{-x}}
\cdot
\frac{e^{-x}}{1+e^{-x}}
$$

Therefore:

$$
\sigma(x)(1-\sigma(x))
=
\frac{e^{-x}}{(1+e^{-x})^2}
$$

This is exactly the derivative we found.

Thus:

$$
\boxed{
\frac{d\sigma(x)}{dx}
=
\sigma(x)(1-\sigma(x))
}
$$

## 2. Derivative of the Tanh Function

The hyperbolic tangent function is defined as:

$$
\tanh(x)
=
\frac{e^x-e^{-x}}{e^x+e^{-x}}
$$

Let:

$$
u=e^x-e^{-x}
$$

and:

$$
v=e^x+e^{-x}
$$

So:

$$
\tanh(x)=\frac{u}{v}
$$

Now compute the derivatives:

$$
u'=e^x+e^{-x}
$$

and:

$$
v'=e^x-e^{-x}
$$

Using the quotient rule:

$$
\left(\frac{u}{v}\right)'
=
\frac{u'v-uv'}{v^2}
$$

Substitute:

$$
\frac{d}{dx}\tanh(x)
=
\frac{
(e^x+e^{-x})(e^x+e^{-x})
-
(e^x-e^{-x})(e^x-e^{-x})
}
{
(e^x+e^{-x})^2
}
$$

This becomes:

$$
\frac{d}{dx}\tanh(x)
=
\frac{
(e^x+e^{-x})^2
-
(e^x-e^{-x})^2
}
{
(e^x+e^{-x})^2
}
$$

Now expand:

$$
(e^x+e^{-x})^2
=
e^{2x}+2+e^{-2x}
$$

and:

$$
(e^x-e^{-x})^2
=
e^{2x}-2+e^{-2x}
$$

Subtracting gives:

$$
(e^{2x}+2+e^{-2x})
-
(e^{2x}-2+e^{-2x})
=
4
$$

Therefore:

$$
\frac{d}{dx}\tanh(x)
=
\frac{4}{(e^x+e^{-x})^2}
$$

Now we show that this equals:

$$
1-\tanh^2(x)
$$

Since:

$$
\tanh^2(x)
=
\left(
\frac{e^x-e^{-x}}{e^x+e^{-x}}
\right)^2
$$

Then:

$$
1-\tanh^2(x)
=
1-
\left(
\frac{e^x-e^{-x}}{e^x+e^{-x}}
\right)^2
$$

Using a common denominator:

$$
1-\tanh^2(x)
=
\frac{
(e^x+e^{-x})^2
-
(e^x-e^{-x})^2
}
{
(e^x+e^{-x})^2
}
$$

From the previous calculation, the numerator is:

$$
4
$$

Therefore:

$$
1-\tanh^2(x)
=
\frac{4}{(e^x+e^{-x})^2}
$$

This is exactly the derivative we found.

Thus:

$$
\boxed{
\frac{d}{dx}\tanh(x)
=
1-\tanh^2(x)
}
$$
